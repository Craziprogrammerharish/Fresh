## Outlier treatment using iqr:

```python
Q3 = df.quantile(0.75)
Q1 = df.quantile(0.25)
IQR = Q3-Q1

df_out_IQR = df[~((df < (Q1 - ( 1.5 * IQR ))) | (df > (Q3 + ( 1.5 * IQR))))]
```
